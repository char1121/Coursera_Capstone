```python
import pandas as pd
```

```python
#import coordinates data
url='https://cocl.us/Geospatial_data/Geospatial_Coordinates.csv'
df_geo=pd.read_csv(url)
df_geo = df_geo.rename(columns = {'Postal Code':'PostalCode'}) 
df_geo.head()
```

```python
#bring in part 1 data
df=pd.read_csv('Capstone.csv')
df.head()
```

```python
#combine coordinates and borough information
df= pd.merge(df, df_geo, on = 'PostalCode')
df.head()
```
